```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Load data
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

```python
# Detect and remove outliers (users with an unusually high number of ratings)
user_counts = ratings['user_id'].value_counts()
threshold = user_counts.quantile(0.99)  # Remove the top 1% of users who rated the most movies
outlier_users = user_counts[user_counts > threshold].index
filtered_ratings = ratings[~ratings['user_id'].isin(outlier_users)]
```

```python
# Update the user-item matrix without outliers
userRatings_filtered = filtered_ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
```

```python
# Train/test split for evaluation
train_ratings, test_ratings = train_test_split(myRatings, test_size=0.2)
```

```python
# Recompute correlation matrix with filtered data
corrMatrix_filtered = userRatings_filtered.corr(method='pearson', min_periods=100)
```

```python
predicted_ratings = {}

for movie in test_ratings.index:
    # Retrieve similar movies from the correlation matrix
    similar_movies = corrMatrix_filtered[movie].dropna()
    
    # Filter to only include movies present in the training set
    similar_movies = similar_movies[similar_movies.index.isin(train_ratings.index)]
    
    if similar_movies.empty:
        # If no similar movies are found, skip the movie
        predicted_ratings[movie] = np.nan
        continue
    
    # Calculate the predicted rating by using the dot product and normalizing by the sum of similarities
    predicted_ratings[movie] = np.dot(similar_movies, train_ratings.loc[similar_movies.index]) / similar_movies.sum()
```

```python
# Evaluate predictions
print("Predicted vs Actual Ratings:")
for movie, predicted_rating in predicted_ratings.items():
    print(f"{movie}: Predicted = {predicted_rating:.2f}, Actual = {test_ratings[movie]:.2f}")
```
